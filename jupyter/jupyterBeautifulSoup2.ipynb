# 웹상의 HTML을 가져오는 방법
## urlopen & BeautifulSoup 사용
### findAll 사용

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup as bs

html = urlopen("http://pythonscraping.com/pages/warandpeace.html")
soup = bs(html, 'html.parser')

nameList = soup.findAll('span', {'class':{'green', 'red'}}) #전체 글 중 span 안에 class가 green과 red를 가져오기
for name in nameList:
    print(name.string)

print("\n")
    
prince = soup.findAll(text = 'The prince') #사용된 단어가 몇번 나왔는지 확인
print("The prine : " ,len(prince))

print("\n")

nameCount = soup.findAll('span', {'class':{'green', 'red'}}, text = 'Heavens! what a virulent attack!')
print("Heavens! what a virulent attack! : " , len(nameCount)) # 정확하게 그 클래스 안에 글이 일치하는 것만 가능하다. Heavens 만쓰면 안나옴

print("\n")

tprince = soup.findAll(text = 'the prince', limit=3) # 여러개의 중복된 값을 limit로 지정가능
for p in tprince:
    print(p)

print("\n")

classRed = soup.findAll(class_ = 'red') # id & class 이름을 가진 태그가 몇개인지 알 수 있다.
print("class = red 의 수 : " ,len(classRed))
```
